```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```


*   Вивести перші рядки таблиці за допомогою методу head



```python
url = "https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8"
tables = pd.read_html(url, match='Коефіцієнт народжуваності в регіонах України', decimal=',', thousands='.')
fert_rate = tables[0]
fert_rate.head()
```

*   Замініть типи нечислових колонок на числові. Підказка - це колонки, де знаходився символ "—"

```python
# fert_rate['2014'] = pd.to_numeric(fert_rate['2014'], errors='coerce')
# fert_rate['2019'] = pd.to_numeric(fert_rate['2019'], errors='coerce')
# fert_rate.dtypes

# Перетворення всіх стовпців, крім 'Регіон', на числові значення
cols_to_convert = fert_rate.columns[1:]  # Все столбцы, кроме первого

# Застосування перетворення до всіх необхідних стовпців
fert_rate[cols_to_convert] = fert_rate[cols_to_convert].apply(pd.to_numeric, errors='coerce')
fert_rate.dtypes
```

*   Видаліть з таблиці дані по всій країні, останній рядок таблиці

```python
fert_rate = fert_rate.drop(index=fert_rate.index[-1])
fert_rate
```

*   Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)

```python

# Конвертація в числовий формат тільки числових стовпців
numeric_columns = fert_rate.columns[1:]
fert_rate[numeric_columns] = fert_rate[numeric_columns].apply(pd.to_numeric, errors='coerce')

# Створення нової таблиці fert_rate_new
fert_rate_new = fert_rate.copy()

# Заміна пропущених значень середніми значеннями за стовпцями в fert_rate_new
fert_rate_new[numeric_columns] = fert_rate_new[numeric_columns].fillna(fert_rate_new[numeric_columns].mean())
fert_rate_new
```

*   Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні

```python
# Обчислення середнього рівня народжуваності у 2019 році
average_birth_rate_2019 = fert_rate_new['2019'].mean()
# average_birth_rate_2019
# print()
# Отримання списку регіонів, де рівень народжуваності у 2019 році був вищим за середній
regions_above_average = fert_rate_new[fert_rate_new['2019'] > average_birth_rate_2019]['Регіон']
regions_above_average
```

*   У  якому регіоні була найвища народжуваність у 2014 році?

```python
max_birth_rate_2014 = fert_rate_new['2014'].idxmax()
region_max_birth_rate_2014 = fert_rate_new.loc[max_birth_rate_2014, 'Регіон']
region_max_birth_rate_2014
```

*   Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році

```python
plt.figure(figsize=(14, 7))

# Створення стовпчастої діаграми
plt.bar(
    fert_rate_new['Регіон'].loc[0:26],  # Мітки стовпців: регіони від 0 до 26
    fert_rate_new['2019'].loc[0:26],    # Висота стовпців: значення народжуваності за 2019 рік для регіонів від 0 до 26
    color=['blue']           # Колір стовпців
)

# Налаштування міток осей та заголовка
plt.xlabel('Регіон')       # Підпис осі X
plt.ylabel('Рівень народжуваності у 2019 році')  # Підпис осі Y
plt.xticks(rotation=90)
plt.legend(['2019 рік'])
plt.title('Діаграма народжуваності по регіонах у 2019 році')  # Заголовок графіка

# Відображення графіка
plt.show()
```

```python
# Побудова лінійного графіка для одного регіону
region = 'Волинська'
plt.figure(figsize=(10, 6))
plt.plot(fert_rate.columns[1:], fert_rate[fert_rate['Регіон'] == region].iloc[0, 1:], marker='o', linestyle='-', color='r')
plt.title(f'Зміна коефіцієнту народжуваності у регіоні {region}')
plt.xlabel('Рік')
plt.ylabel('Рівень коефіцієнту народжуваності')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Побудова гістограми для коефіцієнту народжуваності в 1990 році
plt.figure(figsize=(10, 6))
plt.hist(fert_rate['1990'].dropna(), bins=10, color='y', alpha=0.7)
plt.title('Розподіл коефіцієнту народжуваності в регіонах у 1990 році')
plt.xlabel('Коефіцієнт народжуваності')
plt.ylabel('Частота')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Дані для кругової діаграми коефіцієнт народжуваності у 2000 році
labels = fert_rate['Регіон']
sizes = fert_rate['2000'].fillna(0)  # Заповнюємо NaN нулями

plt.figure(figsize=(10, 10))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Частки коефіцієнту народжуваності у різних регіонах в 2000 році')
plt.axis('equal')  # Забезпечуємо кругову форму діаграми
plt.tight_layout()
plt.show()
```
