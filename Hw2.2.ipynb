# Візуалізація

```python
tmp_visualization = data_file.copy()
for column in tmp_visualization.columns:
  if tmp_visualization[column].isnull().sum() > 0:
    tmp_visualization = tmp_visualization.drop(column, axis=1)
tmp_visualization
```

```python
# Графік розподілу зарплат на місяць
plt.figure(figsize=(10, 6))
sns.histplot(tmp_visualization['Зарплата на місяць'], bins=20, kde=True, color='skyblue')
plt.title('Розподіл зарплат на місяць')
plt.xlabel('Зарплата на місяць')
plt.ylabel('Частота')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Підрахунок кількості працівників за кожною посадою
position_counts = tmp_visualization['Посада'].value_counts().reset_index()
position_counts.columns = ['Посада', 'Кількість']

plt.figure(figsize=(10, 6))

sns.barplot(x='Посада', y='Кількість', hue='Посада', data=position_counts, dodge=False, palette='viridis', legend=False)
plt.title('Кількість працівників за посадами')
plt.xlabel('Посада')
plt.ylabel('Кількість працівників')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Розподіл досвіду роботи на поточному місці роботи
experience_counts = tmp_visualization['Досвід роботи на поточному місці'].value_counts()

# Відфільтруємо дані для показників більше 1.5%
experience_counts_filtered = experience_counts[experience_counts / experience_counts.sum() > 0.015]

plt.figure(figsize=(10, 10))
plt.pie(
    experience_counts_filtered,
    labels=experience_counts_filtered.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=sns.color_palette('pastel'),
    wedgeprops={'edgecolor': 'black', 'linewidth': 1.5},
    pctdistance=0.85,  # Відстань від центру для відсотків
    labeldistance=1.1  # Відстань від центру для підписів
)
plt.title('Розподіл досвіду роботи на поточному місці (більше 1.5%)')
plt.axis('equal')
plt.tight_layout()
plt.show()
```
